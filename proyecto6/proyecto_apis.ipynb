```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
#url
url="https://api.jikan.moe/v4/top/anime"
data=requests.get(url)
if data.status_code==200:
    data=data.json()
    anime_list=[]
    for e in data["data"]:
        anime_info={
            "id": e.get("mail_id"),
            "title": e.get("title"),
            "images":  e.get("images", {}).get("jpg", {}).get("image_url"),
            "rank": e.get("rank"),
            "score": e.get("score"),
        }
        anime_list.append(anime_info)
    df = pd.DataFrame(anime_list).head(10)
    print(df)
else:
    print("Error 404")
```

GRAFICA

```python
plt.figure(figsize=(10,6))
plt.barh(df["title"], df["score"], color="skyblue")
plt.xlabel("score")
plt.ylabel("Anime Title")
plt.title("Top Anime Scores")
plt.gca().invert_yaxis()
plt.show()
```
